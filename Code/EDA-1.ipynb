```python
# import libraries  and packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# About Dataset

* **Artist_name**: The name of the artist

* Track_name: The name of the song

* Release_date: When this song was released

* Genre: The categorical genre of this song

* Lyrics: The pre-tokenized lyrics of this song. Disclaimer: note that as this is real-world data, lyrical content is often obscene. 

* Len:  The number of words in the lyrics of this song

* Dating: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with dating.

* Violence: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with violence.

* World/life: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with the world or life in general terms.

* Night/time: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do night-life or time.

* Shake the audience: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with provocative feeling.

* Family/gospel: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with family-oriented content or the gospel.

* Romantic: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with romantic feeling.

* Communication: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with communication (either in romantic terms or otherwise).

* Obscene: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with obscene content (money, rockstar-lifestyle, etc).

* Music: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with music (music about music, basically).

* Movement/places: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with movement or various locations.

* Light/visual perceptions: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with the sun or other physical weather-related patterns.

* Family/spiritual: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with the importance of family or spirituality.

* Sadness: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with the importance of family or spirituality.

* Feelings: A score from 0 to 1 expressing how likely it is that this song’s lyrics have something to do with emotions, either positive or negative.

* Topic: The categorical label of lyrical content

* Age: A score from 0 to 1 expressing how “old” a song is from our perspective. 1 being the oldest, and 0 being the newest.

```python
plt.figure(figsize=(10, 8))
# Create scatterplot using "dating", "romantic" & "genre" as hue
sns.scatterplot(data=df, x="dating", y="romantic", hue="genre") 
plt.title("Dating vs. Romantic Themes by Genre")
plt.xlabel("Dating Score")
plt.ylabel("Romantic Score")

# Moves legend outside
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left')  
plt.tight_layout()
plt.show()
```

## Dating and Romantic Distribution

* **There's a dense cluster of songs with low dating scores (towards the left side of the plot). This suggests that many songs in the dataset don't heavily focus on dating themes.**

* **Pop songs appear to be distributed across a wider range of both dating and romantic scores, with a noticeable presence in the higher romantic score regions, though not exclusively.**

## Potential Relationships

* **There doesn't appear to be a strong linear relationship between dating and romantic scores across all genres. Some genres might exhibit a slight positive or negative trend if isolated, but overall, the relationship is complex and genre-dependent.**

```python
# Explore various feature vs "release_date" using lineplot 
features = [
    'dating', 'violence', 'world/life', 'night/time', 'shake the audience', 'family/gospel', 
    'romantic', 'communication', 'obscene', 'music', 'movement/places', 'light/visual perceptions', 
    'family/spiritual', 'sadness', 'feelings'
]


fig, axes = plt.subplots(3, 5, figsize=(20, 12))

for idx, feature in enumerate(features):
    row = idx // 5
    col = idx % 5
    ax = axes[row, col]
    

    sns.lineplot(data=df, x="release_date", y=feature, ax=ax, errorbar= None)
    ax.set_title(f'{feature} vs release_date ', fontsize=10)
    ax.set_xlabel("release_date ")
    ax.set_ylabel(feature)

plt.tight_layout()
plt.show()

```

## Observed Trends

* **dating vs release_date: Shows a general downward trend in the 'dating' score over time, with some fluctuations. The score appears to be lower in more recent years compared to earlier periods.**

* **violence vs release_date: Displays an upward trend in the 'violence' score, particularly noticeable from around the mid-20th century onwards. There are fluctuations, but the overall level seems higher in recent years.**

* **romantic vs release_date: Shows a notable decrease from earlier years to more recent ones, indicating a declining trend in the 'romantic' score over time.**

* **movement/places vs release_date: Shows a fluctuating pattern, with a notable increase in the middle period followed by a decrease and then a more recent upward movement.**

* **light/visual perceptions vs release_date: Exhibits a downward trend overall, suggesting a decreasing emphasis on 'light/visual perceptions' over time.**

* **family/spiritual vs release_date: Shows a downward trend, similar to 'family/gospel', indicating a decrease in the prevalence of 'family/spiritual' themes**

## Take away

* **Some features (like 'violence' and 'obscene') show an increasing trend, while others (like "dating", "family/gospel", "romantic", and "music") show a decreasing trend. Many exhibit more complex patterns with fluctuations and periods of relative stability**

# Multivartiate Analysis

```python
#use groupby for "genre"& "topic"
topic_genre_counts_df = df.groupby(['genre', 'topic']).size().reset_index(name='count')

plt.figure(figsize=(14, 8))
sns.barplot(x='genre', y='count', hue='topic', data=topic_genre_counts_df)
plt.title('Topic Frequency by Genre')
```

## Interpreting Topic Frequencies Across Genres

* **"Sadness" is a frequently occurring topic across many genres, particularly blues, country, and rock.**

* **"Romantic" themes are prominent in pop and country music.**

* **"Violence" is a major theme in hip hop and rock.**

* **"World/life" has a strong presence in reggae.**

* **"Music" as a topic appears relatively more in jazz.**

* **The frequency of "obscene" language varies, being more notable in hip hop and rock.**

* **"Feelings" are discussed across genres, with higher counts in blues, country, and pop.**

* **"Night/time" appears to be a less dominant theme across most of these genres in this dataset.**
